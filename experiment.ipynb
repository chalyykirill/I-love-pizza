```python
# парсинг ссылок с https://rutube.ru/video/category/8/13042024/

import requests
from bs4 import BeautifulSoup

# https://rutube.ru/api/comments/video/7dbe8167b0d6bf6330840ebda6615aa4/
# 'https://rutube.ru/video/7dbe8167b0d6bf6330840ebda6615aa4/'



url = 'https://rutube.ru/video/7dbe8167b0d6bf6330840ebda6615aa4/'
# url = 'https://rutube.ru/video/category/8/13042024/'

headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}

def get_html(url):

    r = requests.get(url=url, headers=headers)
    return r.text

def get_total_pages(html):
    soup = BeautifulSoup(html, 'html.parser')
    print(soup.prettify())



get_total_pages(get_html(url))
```
